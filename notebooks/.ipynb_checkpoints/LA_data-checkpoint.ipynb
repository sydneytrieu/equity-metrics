```python
import pandas as pd
import numpy as np
```

```python
HPS_dict = {}

for i in np.arange(1, 17):
    
    if i < 10:
        num_str = '0' + str(i)
    else:
        num_str = str(i)
    
    week = 'week' + num_str
    path = 'C:\\Users\\scher\\data-angels\\data\\pulse2020_puf_'+num_str+'.csv'

    HPS_dict[week] = pd.read_csv(path)
```

```python
HPS_LA = pd.DataFrame()

for df in HPS_dict.values():
    HPS_LA = HPS_LA.append(df[df['EST_MSA']==31080], ignore_index=True)
```

```python
HPS_LA.to_csv(path_or_buf='C:\\Users\\scher\\data-angels\\outputs\\HPS_LA.csv',index=False)
```
